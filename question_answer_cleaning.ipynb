```python
def stage_0_filter(x):
    return len(x.split()) > 30
```

```python
def stage_1_filter(x):
    if '?' in x:
        return True
    return False
```

```python
def stage_2_filter(x):
    wh_words = ['what', 'who', 'where', 'when', 'where', 'why', 'how', 'if', 'have you', 
                'was there', 'is there', 'are there', 'do you', 'is this', 'can you', '']
    x = x.lower()
    for ww in wh_words:
        if ww in x:
            return True
    return False
```

```python
def semifinal_filter(x):
  return stage_0_filter(x) and (stage_1_filter(x) or stage_2_filter(x))
```

```python
def answer_filter(x):
  return len(x.split()) > 100
```

```python
count = 0
for i in range(len(data)):
  pairs = data[i]['Q_A_pairs']
  v_qa_pairs = []
  for pair in pairs:
    if semifinal_filter(pair[0]) and answer_filter(pair[1]):
      v_qa_pairs.append(pair)
  count += len(v_qa_pairs)
  data[i]['Q_A_pairs'] = v_qa_pairs
print(count)
```

```python
valid_qa_pairs = []
for i in range(len(data)):
  for pairs in data[i]['Q_A_pairs']:
    question = pairs[0]
    answer = pairs[1]
    if semifinal_filter(question) & answer_filter(answer):
      valid_qa_pairs.append(pairs)
```

```python
filtered_qa_pairs = []
for i in range(len(valid_qa_pairs)):
  lis = re.split("([?|.] )", valid_qa_pairs[i][0])
  j = 0
  joint_lis = []
  while  j < len(lis)-1:
    joint_lis.append(lis[j] + lis[j+1])
    j = j + 2
  
  for k in joint_lis:
    if semifinal_filter(k):
      #filtered_qa_pairs.append({'question': k, 'answer': valid_qa_pairs[i][0]})
      # print(valid_qa_pairs[i][1])
      # print("===================")
      filtered_qa_pairs.append( [k, [valid_qa_pairs[i][1]] ] )
```
