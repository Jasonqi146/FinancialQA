```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

avg_org_tokens = 0
for i in df['questions']:
  tokens = tokenizer.tokenize(i)
  avg_org_tokens += len(tokens)
avg_org_tokens /= 10
avg_org_tokens
```

```python
print(df['results'])
```

```python
avg_gen_tokens = 0
for i in range(len(df['results'])):
  if pd.isna(df['results'][i]):
    avg_gen_tokens += 1
    continue
  tokens = tokenizer.tokenize(df['results'][i])
  avg_gen_tokens += len(tokens)
avg_gen_tokens /= 10
avg_gen_tokens
```

```python
from tqdm.notebook import tqdm
```

```python
avg_all_q_tokens = 0
count_q = 0
for i in tqdm(range(len(data))):
  doc = data[i]
  for j in range(len(doc['Q_A_pairs'])):
    pair = doc['Q_A_pairs'][j]
    count_q += 1
    tokens = tokenizer.tokenize(pair[0])
    avg_all_q_tokens += len(tokens)
avg_all_q_tokens/count_q
```
